```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv('titanic_train.csv')
df_test = pd.read_csv('titanic_test.csv')
```

```python
from scipy.stats import mode 
mode_embarked = mode(df_train['Embarked'])[0][0]
df_train['Embarked'] = df_train['Embarked'].fillna("S")
df_test['Embarked'] = df_test['Embarked'].fillna("S")
```

```python
df_train['Embarked_C']=df_train['Embarked_C'].astype(int)
df_train['Embarked_S']=df_train['Embarked_S'].astype(int)
df_train['Embarked_Q']=df_train['Embarked_Q'].astype(int)
```

```python
df_test = pd.concat([df_test, pd.get_dummies(df_test['Embarked'], prefix='Embarked')], axis=1)
```

```python
df_test['Embarked_C']=df_test['Embarked_C'].astype(int)
df_test['Embarked_S']=df_test['Embarked_S'].astype(int)
df_test['Embarked_Q']=df_test['Embarked_Q'].astype(int)
```

```python
df_train['gender'] = df_train['Sex'].map({'female':0, 'male':1}).astype(int)
df_test['gender'] = df_test['Sex'].map({'female':0, 'male':1}).astype(int)
```

```python
df_train.dtypes
```

```python
df_train = df_train.drop(['Sex', 'Embarked'], axis =1)
df_test = df_test.drop(['Sex', 'Embarked'], axis =1)
```

```python
df_train.dtypes
```

```python
df_train['family_size'] = df_train['SibSp'] + df_train['Parch'] + 1
df_test['family_size'] = df_test['SibSp'] + df_test['Parch'] + 1
```

```python
def bin_family(x):
    if x>=2 and x <=4 : return 2
    if (x>4 and x<=7) or x==1 : return 1
    if x>7: return 0
```

```python
df_train['family_size'] = df_train['family_size'].apply(bin_family)
df_test['family_size'] = df_test['family_size'].apply(bin_family)
```

```python
df_train.dtypes
```

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
```

```python
df_train.dtypes
```

```python
df_train = df_train.drop(['SibSp','Parch'], axis=1)
df_test = df_test.drop(['SibSp','Parch'], axis=1)
```

```python
def get_title(x):
    return x.split(',')[-1].split('.')[0].strip()
```

```python
df_train['salutation'] = df_train['Name'].apply(get_title)
```

```python
df_test['salutation'] = df_test['Name'].apply(get_title)
```

```python
def salutation_num(x):
    if x in ['Don','Major','Capt','Jonkheer','Rev','Col', 'Sir', 'Master', 'Dr', 'Mr']: return 0
    elif x in ['the Countess','Mme', 'Lady', 'Dona', 'Mrs'] : return 1
    elif x in ['Mlle', 'Ms', 'Miss'] : return 2
    else: return x
#Farhaum
```

```python
df_train['salute'] = df_train['salutation'].apply(salutation_num)
df_test['salute'] = df_test['salutation'].apply(salutation_num)
```

```python
df_train = pd.read_csv('titanic_train_cleaned.csv')
df_test = pd.read_csv('titanic_test_cleaned.csv')
```

```python
def ticket_clean(x):
    x = x.replace('.','')
    x = x.replace('/','')
    x = x.replace(' ','')
    return x
```

```python
df_train['Ticket']
```

```python
df_train['Ticket'] = df_train['Ticket'].apply(ticket_clean)
df_test['Ticket'] = df_test['Ticket'].apply(ticket_clean)
```

```python
df_train['Ticket']
```

```python
ticket_count_train = dict(df_train['Ticket'].value_counts())
ticket_count_test = dict(df_test['Ticket'].value_counts())
```

```python
def ticket_group_train(x):
    return ticket_count_train[x]
def ticket_group_test(x):
    return ticket_count_test[x]
```

```python
df_train['Ticket_Group'] = df_train['Ticket'].apply(ticket_group_train)
df_test['Ticket_Group'] = df_test['Ticket'].apply(ticket_group_test)
```

```python
df_train['Ticket_Group'].unique()
df_test['Ticket_Group'].unique()
```

```python
def ticket_label(x):
    if (x >= 2) & (x <= 4):
        return 2
    elif ((x > 4) & (x <= 8)) | (x == 1):
        return 1
```

```python
df_train['Ticket_lbl'] = df_train['Ticket_Group'].apply(ticket_label)
df_test['Ticket_lbl'] = df_test['Ticket_Group'].apply(ticket_label)
```

```python
df_test['Ticket_lbl'].unique()
```
